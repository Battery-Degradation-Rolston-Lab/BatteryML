```python
import os
import pickle
import pandas as pd
import matplotlib.pyplot as plt

# Directory containing the .pkl files
folder_path = r"C:\Users\rishi\Desktop\batteryml\BatteryML-main\BatteryML-main\processed_data\Custom"

# Initialize an empty list to store all RUL dataframes
all_rul_dfs = []

# End-of-life (EOL) state of health (SOH) threshold
eol_soh = 0.8
min_rul_limit = 100.0
nominal_capacity = 4000  # Updated nominal capacity value

# Iterate through each .pkl file in the folder
for file in os.listdir(folder_path):
    if file.endswith(".pkl"):
        file_path = os.path.join(folder_path, file)
        
        # Load the .pkl file
        with open(file_path, 'rb') as f:
            cell_data = pickle.load(f)
        
        # Initialize lists to store cycle numbers and RULs
        cycle_numbers = []
        ruls = []
        
        # Determine the RUL for each cycle
        found_eol = False
        for i, cycle in enumerate(cell_data['cycle_data']):
            cycle_number = cycle['cycle_number']
            cycle_numbers.append(cycle_number)
            
            # Calculate maximum discharge capacity for the cycle
            Qd = max(cycle['discharge_capacity_in_Ah'])
            
            # Check if EOL condition is met
            if not found_eol and Qd <= nominal_capacity * eol_soh:
                rul = len(cell_data['cycle_data']) - i
                found_eol = True
            else:
                rul = float('nan')
            
            # Apply min RUL limit
            if rul <= min_rul_limit:
                rul = float('nan')
                
            ruls.append(rul)
            
            # Debug: print the calculated Qd and eol_soh condition
            print(f"Cycle {cycle_number}: Qd={Qd}, EOL Condition Met={Qd <= nominal_capacity * eol_soh}")
        
        # Debug: print the processed data for each file
        print(f"Processed {file}:")
        print(f"Cycle Numbers: {cycle_numbers}")
        print(f"RULs: {ruls}")
        
        # Create a DataFrame for plotting
        df_rul = pd.DataFrame({'cycle number': cycle_numbers, 'RUL': ruls})
        
        # Add the DataFrame to the list
        all_rul_dfs.append((os.path.basename(file), df_rul))

# Plotting all data on the same graph
plt.figure(figsize=(12, 8))  # Increase figure size for better visibility
for filename, df in all_rul_dfs:
    plt.plot(df['cycle number'], df['RUL'], marker='o', linestyle='-', label=filename)

plt.title('Cycle Number vs Remaining Useful Life (RUL) for All Cells')
plt.xlabel('Cycle Number')
plt.ylabel('Remaining Useful Life (RUL)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import os
import pickle

# Replace with the path to your directory containing pkl files
directory_path = r'C:\Users\rishi\Desktop\batteryml\BatteryML-main\BatteryML-main\processed_data'

def calculate_total_cycles(file_path):
    with open(file_path, 'rb') as file:
        data = pickle.load(file)
    
    # Assuming the structure of the data, cycle data is stored in `cycle_data`
    if 'cycle_data' in data:
        total_cycles = len(data['cycle_data'])
        return total_cycles
    else:
        print(f"The key 'cycle_data' does not exist in the data for file {file_path}.")
        return None

# Iterate over all files in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.pkl'):
        file_path = os.path.join(directory_path, filename)
        total_cycles = calculate_total_cycles(file_path)
        if total_cycles is not None:
            print(f"Total number of cycles in {filename}: {total_cycles}")
```

```python
import pandas as pd

# Define the columns based on the provided column names
columns = [
    'mode', 'ox/red', 'error', 'control changes', 'Ns changes', 'counter inc.',
    'Ns', 'I Range', 'time/s', 'control/V/mA', 'Ecell/V', 'I/mA', 'dq/mA.h', 
    '(Q-Qo)/mA.h', '|Energy|/W.h', 'freq/Hz', '|Z|/Ohm', 'Phase(Z)/deg', 
    'Q charge/discharge/mA.h', 'half cycle', 'Energy charge/W.h', 'Energy discharge/W.h',
    'Capacitance charge/F', 'Capacitance discharge/F', 'step time/s', 'z cycle', 
    'Re(Z)/Ohm', '-Im(Z)/Ohm', 'Re(Y)/Ohm-1', 'Im(Y)/Ohm-1', '|Y|/Ohm-1', 
    'Phase(Y)/deg', 'x', 'Q discharge/mA.h', 'Q charge/mA.h', 'Capacity/mA.h', 
    'Efficiency/%', 'control/V', 'control/mA', 'cycle number', 'P/W', 'R/Ohm'
]

# Function to read and parse the MPT file starting from row 99
def read_mpt_file(file_path):
    data = []
    with open(file_path, 'r') as file:
        for _ in range(98):  # Skip the first 98 rows
            next(file)
        for line in file:
            values = line.strip().split('\t')  # Assuming tab-separated values
            data.append(values)
    return data

# Convert the parsed data to a pandas DataFrame
def convert_to_dataframe(data):
    df = pd.DataFrame(data, columns=columns)
    return df

# Save the DataFrame to an Excel file
def save_to_excel(df, excel_file_path):
    df.to_excel(excel_file_path, index=False)

# Replace with the path to your MPT file
file_path = r'C:\Users\rishi\Desktop\Glimpse\EVE\2407100001Glimpse_CA1.mpt'
data = read_mpt_file(file_path)
df = convert_to_dataframe(data)

# Replace with the desired output Excel file path
excel_file_path = r'C:\Users\rishi\Desktop\dataset custom\mpt to excel.xlsx'
save_to_excel(df, excel_file_path)

print(f"Data successfully converted to {excel_file_path}")
```
